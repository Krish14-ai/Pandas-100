## 3. GroupBy & Aggregation[GroupBy docs](https://pandas.pydata.org/docs/user_guide/groupby.html)

`groupby` lets you split data into groups, apply a function to each, and combine the results.**Worked example:** average revenue by a category column.

```python
sales = pd.DataFrame({    'region': ['North', 'South', 'North', 'East', 'South', 'East'],    'revenue': [120, 90, 150, 60, 110, 75]})sales.groupby('region')['revenue'].mean()
```

## 4. Merging, Joining & Reshaping[Merge, join, concatenate docs](https://pandas.pydata.org/docs/user_guide/merging.html)

Combining datasets is one of the most common real-world tasks: stacking rows, joining on keys, or reshaping between wide and long formats.**Worked example:** merge two DataFrames on a key column.

```python
customers = pd.DataFrame({'cust_id': [1, 2, 3], 'name': ['Ana', 'Bo', 'Cy']})orders = pd.DataFrame({'cust_id': [1, 1, 2], 'amount': [50, 20, 75]})merged = customers.merge(orders, on='cust_id', how='left')merged
```

## 5. String Methods & Time Series[Time series docs](https://pandas.pydata.org/docs/user_guide/timeseries.html) · [String methods docs](https://pandas.pydata.org/docs/user_guide/text.html)

The `.str` accessor vectorizes string operations on text columns, and Pandas has rich support for dates, resampling, and rolling windows.**Worked example:** parse dates and resample to monthly totals.

```python
ts = pd.DataFrame({    'date': pd.date_range('2026-01-01', periods=90, freq='D'),    'value': np.random.randn(90).cumsum()}).set_index('date')ts.resample('ME').mean().head()
```
